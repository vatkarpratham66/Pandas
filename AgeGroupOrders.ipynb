```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r'C:\Users\PRATHAMESH\OneDrive\Pictures\Camera Roll\library\numpydemo\Diwali.csv'
df = pd.read_csv(file_path, encoding='latin1')


gender_orders = df.groupby('Age Group')['Orders'].sum()
print(gender_orders)

colors = [
    'green' if value > 5000 else 'blue' if value > 1000 else 'red'
    for value in gender_orders
]
plt.bar(gender_orders.index, gender_orders.values, color=colors, width=0.4)
plt.title('Number of Orders by Age_group')
plt.xlabel('Age_group')
plt.ylabel('Number of Orders')
#plt.grid(True)
plt.show()
```
