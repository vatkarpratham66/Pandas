```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file path and read CSV
file_path = r'C:\Users\PRATHAMESH\OneDrive\Pictures\Camera Roll\library\numpydemo\Diwali.csv'
df = pd.read_csv(file_path, encoding='latin1')

# Get the gender distribution
value_counts = df['Gender'].value_counts()
print(value_counts)

# Plot the bar chart
plt.bar(value_counts.index, value_counts.values, color=['green', 'red'], width=0.4)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add annotations on the bars
for i, value in enumerate(value_counts.values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom')

# Adjust layout and display
plt.tight_layout()
plt.show()

```
