```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r'C:\Users\PRATHAMESH\OneDrive\Pictures\Camera Roll\library\numpydemo\Diwali.csv'
df = pd.read_csv(file_path, encoding='latin1')


gender_orders = df.groupby('Gender')['Orders'].sum()
print(gender_orders)


plt.bar(gender_orders.index, gender_orders.values, color=['green', 'red'], width=0.4)
plt.title('Number of Orders by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Orders')


plt.show()
```
