```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r'C:\Users\PRATHAMESH\OneDrive\Pictures\Camera Roll\library\numpydemo\Diwali.csv'
df = pd.read_csv(file_path, encoding='latin1')

gender_orders = df.groupby('Occupation')['Orders'].sum()
print(gender_orders)

colors = [
    'green' if value > 2500 else 'blue' if value > 1000 else 'red'
    for value in gender_orders
]
plt.figure(figsize=(20, 6))
plt.bar(gender_orders.index, gender_orders.values, color=colors, width=0.4)
plt.title('Number of Orders by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
#plt.grid(True)
plt.show()
```
