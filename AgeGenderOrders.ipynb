```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r'C:\Users\PRATHAMESH\OneDrive\Pictures\Camera Roll\library\numpydemo\Diwali.csv'
df = pd.read_csv(file_path, encoding='latin1')


gender_orders = df.groupby(['Age Group', 'Gender'])['Orders'].sum().unstack()

print(gender_orders)


plt.figure(figsize=(20, 6))

gender_orders.plot(kind='bar',width=0.6)



plt.title('Number of Orders by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Orders')


plt.tight_layout()
plt.show()
```
